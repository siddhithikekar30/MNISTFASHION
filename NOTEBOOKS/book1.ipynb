```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical 
```

```python
data_train= pd.read_csv('/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/fashion-mnist_train.csv')
data_test = pd.read_csv('/Users/mr.engineer/Desktop/Code/DSPROJECT/MNIST_FASHION/DATA/fashion-mnist_test.csv')

X_train = data_train.drop(columns='label')
y_train = data_train['label'].values
X_test = data_test.drop(columns='label')
y_test = data_test['label'].values
sample = X_train.sample(1).values.reshape(28,28)

sns.heatmap(sample)
```

```python
sns.histplot(y_train)
plt.title("Class Distribution")
plt.show()
```

```python
X_train.describe()
```

```python
#Scaling MIN MAX
X_train = X_train/255.0
X_train.describe()
```

```python
X_train = X_train.values.reshape(-1,28,28,1)
y_cat = to_categorical(y_train,num_classes=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D, BatchNormalization, Flatten
model = Sequential([
    Conv2D(32,(3,3), activation = 'relu', input_shape = (28,28,1)),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(64,(3,3), activation = 'relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(256,activation = 'relu'),
    Dropout(0.3),
    Dense(10,activation = 'softmax')

])
```

```python
model.compile(optimizer='adam',loss = 'categorical_crossentropy', metrics = ['accuracy'])

model.fit(X_train, y_cat, epochs = 15, batch_size = 32, validation_split = 0.2)
```

```python
y_test_cat = to_categorical(y_test,num_classes=10)
X_test = X_test/255.0
X_test = X_test.values.reshape(-1,28,28,1)
test_loss, test_acc = model.evaluate(X_test,y_test_cat)
print("Test Accuracy: ", test_acc)
```
